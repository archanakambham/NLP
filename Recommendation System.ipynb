```python
import pandas as pd
review=pd.read_csv('Hotel_Reviews.csv')
review.head()
```

```python
s=[]
for i in review['Average_Score']:
    if i>=1 and i<=3:
        s.append('Bad')
    elif i>=4 and i<=6:
        s.append('Average')
    elif i>7 and i<=8:
        s.append('good')
    else:
        s.append('excellent')
```

```python
review['Hotel_Rating']=s
```

```python
review['Hotel_Rating'].unique()
```

```python
review['city'] = review.Hotel_Address.apply(lambda x: x.split(' ')[-2])
review['country'] = review.Hotel_Address.apply(lambda x: x.split(' ')[-1])
```

```python
neg_rev= review['Negative_Review']
pos_rev = review['Positive_Review']
```

```python
leisure = review['Tags'].str.contains(' Leisure trip ')
business = review['Tags'].str.contains(' Business trip ')
```

```python
data = pd.DataFrame({'Combined_Review': neg_rev + pos_rev, 'Trip_Type': leisure})
data.replace({True: 'Leisure', False: 'Business'}, inplace = True)
data[20:30]
```

```python
a=pd.concat([review,data],axis=1)
a.head()
```

```python
import nltk
import re
from nltk.corpus import stopwords
from sklearn.metrics.pairwise import linear_kernel
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
a=a[a['Hotel_Name'].notna()]
a['aggregated_text']=a['city'].str.lower()+a['Trip_Type'].str.lower()+a['Hotel_Rating'].str.lower()
corpus=list(a['aggregated_text'].str.split())
stopwords_list=set(stopwords.words("english"))
index=list(range(0,len(corpus)))
clean_corpus=[]

for sentence in corpus:
    s=[]
    for word in sentence:
        clean_word=re.sub(r'[^\w\s]','', word)
        if clean_word not in stopwords_list:
            s.append(clean_word)
        clean_corpus.append(" ".join(s)) 
```

```python
tfidf_vectorizer = TfidfVectorizer().fit_transform(clean_corpus)
```

```python
def get_recommendation(show_list, vectorizer):

    Hotel_Name, scores, city = [], [], []
    for show_name in show_list:
        show_index = a[a['Hotel_Name'] == show_name].index[0]
        cosine_similarities = linear_kernel(vectorizer[show_index], vectorizer).flatten()
        similar_show_index = cosine_similarities.argsort()[1:]
        Hotel_Name += [a['Hotel_Name'][i] for i in similar_show_index]
        city += [a['city'][i] for i in similar_show_index]
        scores += list(cosine_similarities[similar_show_index])

    df = pd.DataFrame(data = {'Hotel_Name': Hotel_Name, 
                                 'city': city,'Cosine_similarity': scores})

    df = df[~df['Hotel_Name'].isin(show_list)].sort_values('Cosine_similarity', ascending = False)
    df['Hotel_Name'] = df['Hotel_Name'].drop_duplicates()
    top_five_list = df[df['Hotel_Name'].notna()].iloc[0:5, :]
    
    return top_five_list
```

```python
Hotel_name = ['Hotel Arena']
get_recommendation(Hotel_name, tfidf_vectorizer)
```
